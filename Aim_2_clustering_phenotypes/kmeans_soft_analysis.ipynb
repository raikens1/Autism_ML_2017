```python
##############
## Clusters ##
##############

X2 = pd.read_csv('/scratch/PI/dpwall/DATA/iHART/kpaskov/PhenotypeGLRM/data/softkmeans_logloss_X2_10.24.17.csv', \
                 index_col=0, header=None)
X3 = pd.read_csv('/scratch/PI/dpwall/DATA/iHART/kpaskov/PhenotypeGLRM/data/softkmeans_logloss_X3_10.24.17.csv', \
                 index_col=0, header=None)
X4 = pd.read_csv('/scratch/PI/dpwall/DATA/iHART/kpaskov/PhenotypeGLRM/data/softkmeans_logloss_X4_10.24.17.csv', \
                 index_col=0, header=None)
X5 = pd.read_csv('/scratch/PI/dpwall/DATA/iHART/kpaskov/PhenotypeGLRM/data/softkmeans_logloss_X5_10.24.17.csv', \
                 index_col=0, header=None)
X6 = pd.read_csv('/scratch/PI/dpwall/DATA/iHART/kpaskov/PhenotypeGLRM/data/softkmeans_logloss_X6_10.24.17.csv', \
                 index_col=0, header=None)
clusters = [X2, X3, X4, X5, X6]

###############
## Centroids ##
###############

Y2 = pd.read_csv('/scratch/PI/dpwall/DATA/iHART/kpaskov/PhenotypeGLRM/data/softkmeans_logloss_Y2_10.24.17.csv')
Y3 = pd.read_csv('/scratch/PI/dpwall/DATA/iHART/kpaskov/PhenotypeGLRM/data/softkmeans_logloss_Y3_10.24.17.csv')
Y4 = pd.read_csv('/scratch/PI/dpwall/DATA/iHART/kpaskov/PhenotypeGLRM/data/softkmeans_logloss_Y4_10.24.17.csv')
Y5 = pd.read_csv('/scratch/PI/dpwall/DATA/iHART/kpaskov/PhenotypeGLRM/data/softkmeans_logloss_Y5_10.24.17.csv')
Y6 = pd.read_csv('/scratch/PI/dpwall/DATA/iHART/kpaskov/PhenotypeGLRM/data/softkmeans_logloss_Y6_10.24.17.csv')
centroids = [Y2, Y3, Y4, Y5, Y6]

##############
## Features ##
##############

raw_data = pd.read_csv('/scratch/PI/dpwall/DATA/iHART/kpaskov/PhenotypeGLRM/data/all_samples.csv', \
                       index_col='identifier', \
                       na_values=['None', 'NaN', 'Unknown (not reported)', 'Unknown (not reported) or Other'])
features = format_features(raw_data, Y2.columns, X2.index)

############
## Labels ##
############

raw_data = pd.read_csv('/scratch/PI/dpwall/DATA/iHART/kpaskov/PhenotypeGLRM/data/all_samples.csv', \
                       index_col='identifier', \
                       na_values=['None', 'NaN', 'Unknown (not reported)', 'Unknown (not reported) or Other'])
labels = format_labels(raw_data, X2.index)
```

### feature heatmap

```python
k_values = [2, 3, 4, 5, 6]
single_clusters = [soft_to_crisp_clusters(cluster_result, rule='single') for cluster_result in clusters]
for k in k_values:
    cluster_result = clusters[k-2]
    single_cluster = single_clusters[k-2]
    heatmap_pipeline(features, sort_by=[], clusters=single_cluster, size=(15, 9))
```
